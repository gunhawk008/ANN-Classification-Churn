```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
from sklearn.preprocessing import OneHotEncoder
```

```python
# load the dataset
data=pd.read_csv("Churn_modelling.csv")
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

```python
#Gender Col label encoding into 0 and 1
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
```

```python
onehot_encode_geography = OneHotEncoder()
geo_encoder = onehot_encode_geography.fit_transform(data[['Geography']]).toarray()
geo_encoder
```

```python
onehot_encode_geography.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder,columns=onehot_encode_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
data = pd.concat([data,geo_encoded_df],axis=1)
data = data.drop(['Geography'],axis = 1)
```

```python
data.head()
```

```python
with open("label_encoder_gender",'wb') as file:
    pickle.dump(label_encoder_gender,file)

with open("onehot_encode_geography",'wb') as file:
    pickle.dump(onehot_encode_geography,file)
```

```python
#Divide in dep and indep features

X=data.drop(['Exited'],axis=1)
Y = data['Exited']



```

```python
#Split the data in training and tetsing sets
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=45)

#Scale all the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
with open("scaler","wb") as file:
    pickle.dump(scaler,file)
```

ANN Model Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
print(X_train[:2])
```

```python
# Actual Model Generation

model= Sequential ([
    Dense(64,activation='relu',input_shape=((X_train.shape[1],))), #Hidden Layer 1
    Dense(32,activation='relu'), #Hidden Layer 2
    Dense(1,activation='sigmoid') #Output Layer
])
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
loss = tf.keras.losses.BinaryCrossentropy()
loss
```

```python
model.compile(
    optimizer = opt,
    loss=loss,
    metrics = ['accuracy']
)
```

```python
# Set up tensor Board

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

tf_callback = TensorBoard(log_dir=log_dir,histogram_freq= 1) 
```

```python
early_stopping_callback=EarlyStopping(
    monitor='val_loss',
    patience = 10,
    restore_best_weights=True
)
```

```python
#Train the model

history = model.fit(
    X_train,
    Y_train,
    validation_data = (X_test,Y_test),
    epochs = 100,
    callbacks = [tf_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
